Lexicométrie avec fréquence, puis clustering des questions (avec les réponses) par rapport au type d'entité, etiquetages automatique avec edsnlp

```python
import glob
import json
import spacy
import re
from collections import OrderedDict



def lire_fichier (chemin):
    with open(chemin) as json_data: 
        dist =json.load(json_data)
    return dist
        
def nom_fichier(chemin):
    for mot in glob.glob(chemin): 
        noms_fichiers = re.split("/", chemin)
#        print("NOM FICHIER",noms_fichiers)
        
        nomsfich = re.split("\.",  noms_fichiers[3])
#        print(nomsfich)
    return nomsfich
        
def stocker( chemin, contenu):

    w =open(chemin, "w")
    w.write(json.dumps(contenu , indent = 2))
    w.close()
    print(chemin)
    
    return chemin
    
   
```

```python
#Tokenisation

path="../DATA/train_court.json" 
#for chemin in glob.glob("%s/*.json"%path_corpora):
path_dist=lire_fichier(path)
# print(path_dist)

liste_question=[]
liste_reponse=[]
liste_subject=[]
for l in path_dist:
    # print(l)
    for key, value in l.items():
        # print (value)
        if key=="question":
            liste_question.append(value)
            
            # print("????????valeur question????????",value)
        if key=="answers":
            liste_reponse.append(value)
            # print("---------valeur réponse---------",value)
        if key=="subject_name":
            liste_subject.append(value)
            # print("**********subject_name*******", value)



nlp = spacy.load("fr_core_news_lg")
dico_quest={}   
liste_tok_q=[] 
liste_tok_r=[] 
liste_tok=[]      
for q in liste_question:
    doc = nlp(q)
    for token in doc:
        liste_tok_q.append(token.text)

for r in liste_reponse:
    for keys, values in r.items():
        
        doc = nlp(values)
        for token in doc:
            liste_tok_r.append(token.text)

liste_tok.append(liste_tok_q)
print(len(liste_tok_q))
liste_tok.append(liste_tok_r)
print(len(liste_tok_r))
```

```python
#fréquence
dic_mots_q={}  
i=0  

for mot in liste_tok_q:
    if len(mot) > 3:
        if mot not in dic_mots_q:
            dic_mots_q[mot] = 1
        else:
            dic_mots_q[mot] += 1 
    
new_q = OrderedDict(sorted(dic_mots_q.items(), key=lambda t: t[1]))

print("Fréquence terme Question",new_q)

dic_mots_r={}  
for mot in liste_tok_r:
    if len(mot) > 3:
        if mot not in dic_mots_r:
            dic_mots_r[mot] = 1
        else:
            dic_mots_r[mot] += 1 
new_r = OrderedDict(sorted(dic_mots_r.items(), key=lambda t: t[1]))

print("Fréquence terme Réponse",new_r)
```
